# Problem 4

Given $f(x) = x^4 + 2x^2 - x - 3$, and that $f(p) = 0$, then:

$g_1:$

\begin{gather*}

x = (3 + x - 2x^2)^{\frac{1}{4}} \\ 
x^4 = 3 + x - 2x^2 \\ 
x^4 + 2x^2 - x - 3 = 0 \\
\text{If } f(p) = 0, \text{ then,} \\ 
p^4 + 2p^2 - p - 3 = 0 \\
f(p) = 0 \checkmark

\end{gather*}

$g_2:$

\begin{gather*}

x = (\frac{x + 3 - x^4}{2})^{1/2}\\ 
x^2 = \frac{x + 3 - x^4}{2} \\ 
2x^2 = x + 3 - x^4 \\ 
x^4 + 2x^2 - x - 3 = 0 \\
\text{If } f(p) = 0, \text{ then,} \\ 
p^4 + 2p^2 - p - 3 = 0 \\
f(p) = 0 \checkmark

\end{gather*}

$g_3:$

\begin{gather*}

x = (\frac{x + 3}{x^2 + 2})^{\frac{1}{2}} \\
x^2 = \frac{x + 3}{x^2 + 2} \\ 
x^4 + 2x^2 = x + 3 \\
x^4 + 2x^2 - x - 3 = 0 \\
\text{If } f(p) = 0, \text{ then,} \\ 
p^4 + 2p^2 - p - 3 = 0 \\
f(p) = 0 \checkmark

\end{gather*}

$g_4:$

\begin{gather*}

x = \frac{3x^4 + 2x^2 + 3}{4x^3 + 4x - 1} \\
4x^4 + 4x^2 - x = 3x^4 + 2x^2 + 3 \\ 
x^4 + 2x^2 - x - 3 = 0 \\
\text{If } f(p) = 0, \text{ then,} \\ 
p^4 + 2p^2 - p - 3 = 0 \\
f(p) = 0 \checkmark

\end{gather*}

```python
import numpy as np
import pandas as pd

# make x vals based on n = 8, h = 0.25
n = 1
x = np.linspace(0,2,n+1)
a = 0
b = 2
h = (b-a)/n

f = lambda x : np.e**(2*x) * np.sin(3*x)

fexact = (2*np.e**4*np.sin(6) - 3*np.e**4*np.cos(6) + 3)/(13)

# a. trapezoidal
def compTrap(x, f, h):
    return (h/2)*(f(x[0]) + 2*(sum(f(x[1:len(x)-1]))) + f(x[len(x)-1]))

# b. simpson
def compSimp(x,f,h):
    # extract indices from index 1 to (n/2)-1
    # first sum
    sum1 = 0
    for i in np.arange(1,int((len(x)-1)/2)):
        sum1 += f(x[2*i])
    # extract indices from index 1 to (n/2)
    # second sum
    sum2 = 0
    for i in np.arange(1,int((len(x)-1)/2)+1):
        sum2 += f(x[2*i - 1])

    return (h/3)*(f(x[0]) + 2*sum1 + 4*sum2 + f(x[len(x)-1]))

# c. midpoint
def compMid(x,f,h):
    # extract indices from index 0 to (n/2)
    sum = 0
    for i in np.arange(0,int((len(x)-1)/2)+1):
        sum += f(x[2*i])

    return 2 * h * sum

# check n for trap
while abs(compTrap(x,f,h) - fexact) >= 10**-4:
    n += 1
    x = np.linspace(0,2,n+1)
    h = (b-a)/n
print(compTrap(x,f,h))
print("Composite Trapezoid reaches error tolerance with n = ",n)

# reset for comp simp
n = 2
# check n for simp
while abs(compSimp(x,f,h) - fexact) >= 10**-4:
    n += 2
    x = np.linspace(0,2,n+1)
    h = (b-a)/n
print(compSimp(x,f,h))
print("Composite Simpson's Rule reached error tolerance with n = ",n)

# reset for comp mid
n = 2
h = (b-a)/(n+2)
x = np.linspace(0,2,n+3)
# check n for mid
while abs(compMid(x,f,h) - fexact) >= 10**-4:
    n += 2
    h = (b-a)/(n+2)
    x = np.linspace(0,2,n+3)
    print(compMid(x,f,h))
print(compMid(x,f,h))
print("Composite Midpoint Rule reached error tolerance with n = ",n)
```

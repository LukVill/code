## c.

The error form of the Lagrange polynomial is:

$f(x) = P_n(x) + \frac{f^{(n+1)}\epsilon(x)}{(n+1)!} \Pi_{i = 0}^{n}(x - x_i)$

such that $\epsilon(x) \in (min(x_n), max(x_n))$


So,

\begin{gather*}

f(x) = \ln{x} \\

\implies \\

f^{(4)} = \frac{-6}{x^4} \\

\implies \\

| f(x) - P_3(x) | = \frac{\frac{-6}{\epsilon(x)^4}}{4!}(x-1)(x-1.1)(x-1.3)(x-1.4) \\
\implies \frac{-1}{4\epsilon(x)^4}(x-1)(x-1.1)(x-1.3)(x-1.4)\\

\text{Since } \frac{1}{4*1^4} > \frac{1}{4*1.4^4}, \text{ set } \epsilon(x) = 1 \\

\implies \frac{-1}{4}(x-1)(x-1.1)(x-1.3)(x-1.4) \\

\text{Let } g(x) = (x-1)(x-1.1)(x-1.3)(x-1.4) \\

\text{Since } \forall x \in [1,1.4], \ln{1.4} > \ln{x}, \implies g(x) \leq \ln{1.4} \\

\implies \\

| f(x) - P_3(x) | \leq | \frac{\ln{1.4}}{4} |

\end{gather*}

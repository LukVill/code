```python
import numpy as np

# UPDATED TO ROUND TO 2 DECIMALS
def upperTri(mat):
    # for each column
    for col in range(mat.shape[1]):
        i = col

        # for each row 
        while i < mat.shape[0]-1:
            # check if non zero
            if(mat[i+1,col] != 0):
                # get scalar amount
                c = np.round(-mat[i+1,col]/mat[col,col],2)
                # make multiplied row
                addRow = np.round(mat[col,:] * c,2)
                # add to row
                mat[i+1,:] = np.round(mat[i+1,:] + addRow,2)
            
            # increment row
            i += 1
            
    return mat

# UPDATED TO ROUND TO 2 DECIMALS
# assume that mat and b are already separated
def backsub(mat,b):
    # make result vec
    x = b

    # start i at last index
    i = len(mat)-1
    
    while i >= 0:
        # get x
        # for each entry BUT on curr,
        for j in range(mat.shape[1]):
            if i != j:             
                # subtract x by entries
                x[i] = np.round(x[i] - mat[i,j],2)
        # divide the curr entry
        x[i] = np.round(x[i]/mat[i,i],2)

        # update A with new found variable
        mat[:,i] = np.round(mat[:,i] * x[i],2)

        # decrement
        i -= 1

    return x


def solve(A, b):

    # type cast first
    A = np.float64(A)
    b = np.float64(b)

    # get upper triangle
    matA = np.c_[A,b]
    matA = upperTri(matA)

    # extract A and b
    A = matA[:,:len(matA)]
    b = matA[:,len(matA)]

    # back substitute
    return backsub(A,b)

A = np.matrix([[-1,4,1],[5/3,2/3,2/3],[2,1,4]])
b = [8,1,11]

solve(A,b)
```

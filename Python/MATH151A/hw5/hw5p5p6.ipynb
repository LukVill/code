# Problem 5

$\renewcommand{\R}{\mathbb{R}}$

Expanding $(x_0 + h)$ and $(x_0 + 2h)$ by Taylor Expansion:

Given $h \in \R$:

$$

f(x_0 + h) = f(x_0) + f'(x_0)h + \frac{f''(x_0)}{2}h^2 + \frac{f'''(\epsilon_1)}{6}h^3 \\ \implies

4f(x_0 + h) = 4f(x_0) + f'(x_0)4h + f''(x_0)2h^2 + \frac{4f'''(\epsilon_1)}{6}h^3 \\\\

f(x_0 + 2h) = f(x_0) + f'(x_0)2h + \frac{f''(x_0)}{2}2h^2 + \frac{f'''(\epsilon_2)}{6}8h^3 \\\\

4f(x_0 + h) - f(x_0 + 2h) = 3f(x_0) + f'(x_0)2h + \frac{2f'''(\epsilon_1) - 4f'''(\epsilon_2)}{3}h^3 \\\\

f'(x_0) = \frac{1}{2h}(-3f(x_0) + 4f(x_0 + h) - f(x_0 + 2h)) + \frac{2f'''(\epsilon_2) - f'''(\epsilon_1)}{3}h^2

$$

Let $g(\epsilon_3) = \frac{2f'''(\epsilon_2) - f'''(\epsilon_1)}{3}$. 

Because $f$ is continuous, $\exists M$ such that $\forall x \in \R, |f(x)| \leq M$. 

This implies that $g(x) \leq M$. So, 

$$
f'(x_0) \leq \frac{1}{2h}(-3f(x_0) + 4f(x_0 + h) - f(x_0 + 2h)) + Mh^2
$$

Since $M \in \R$, this implies

$$
f'(x_0) = \frac{1}{2h}(-3f(x_0) + 4f(x_0 + h) - f(x_0 + 2h)) + O{h^2}
$$

Since f'(x) is used, we will use f''(x) as the error term.

\begin{gather*}
f'(x) = cos(x)\\
f''(x) = -sin(x)\\
\end{gather*}

Let $x = x_0 + h$, so

\begin{gather*}
R_2(x) = \frac{f''(x_0)}{2}h^2
\implies \\ Error = |\frac{f''(x_0)}{2}h^2|
\end{gather*}

We will use every point that has a next point and calculate that error approximation.

\begin{gather*}
x_0 \rightarrow 0.5 \\\\
|\frac{sin(0.5)}{2}0.1^2| \\
\approx 0.00239713 \\

x_0 \rightarrow 0.6 \\\\

\approx 0.00282321 \\
\end{gather*}
